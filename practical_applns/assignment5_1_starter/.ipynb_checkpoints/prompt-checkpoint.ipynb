### Required Assignment 5.1: Will the Customer Accept the Coupon?

**Context**

Imagine driving through town and a coupon is delivered to your cell phone for a restaurant near where you are driving. Would you accept that coupon and take a short detour to the restaurant? Would you accept the coupon but use it on a subsequent trip? Would you ignore the coupon entirely? What if the coupon was for a bar instead of a restaurant? What about a coffee house? Would you accept a bar coupon with a minor passenger in the car? What about if it was just you and your partner in the car? Would weather impact the rate of acceptance? What about the time of day?

Obviously, proximity to the business is a factor on whether the coupon is delivered to the driver or not, but what are the factors that determine whether a driver accepts the coupon once it is delivered to them? How would you determine whether a driver is likely to accept a coupon?

**Overview**

The goal of this project is to use what you know about visualizations and probability distributions to distinguish between customers who accepted a driving coupon versus those that did not.

**Data**

This data comes to us from the UCI Machine Learning repository and was collected via a survey on Amazon Mechanical Turk. The survey describes different driving scenarios including the destination, current time, weather, passenger, etc., and then ask the person whether he will accept the coupon if he is the driver. Answers that the user will drive there ‘right away’ or ‘later before the coupon expires’ are labeled as ‘Y = 1’ and answers ‘no, I do not want the coupon’ are labeled as ‘Y = 0’.  There are five different types of coupons -- less expensive restaurants (under \$20), coffee houses, carry out & take away, bar, and more expensive restaurants (\$20 - $50).

**Deliverables**

Your final product should be a brief report that highlights the differences between customers who did and did not accept the coupons.  To explore the data you will utilize your knowledge of plotting, statistical summaries, and visualization using Python. You will publish your findings in a public facing github repository as your first portfolio piece.

**Investigating the Bar Coupons**

Now, we will lead you through an exploration of just the bar related coupons.  

1. Create a new `DataFrame` that contains just the bar coupons.


```python
bar_data = data.query('coupon == "Bar"')
bar_data.shape
```

2. What proportion of bar coupons were accepted?


```python
print(bar_data['Y'].sum())
print(str(bar_data['Y'].value_counts(0)[1]) + ' drivers chose to accept the coupon (' + str(bar_data['Y'].value_counts(1)[1]*100) + '%)')
```

3. Compare the acceptance rate between those who went to a bar 3 or fewer times a month to those who went more.


```python
less_than_3times_list = ['never','less1','1~3']

less_than_3times_accept_rate = bar_data.query("Bar in @less_than_3times_list")['Y'].value_counts(1)[1]*100
more_than_3times_accept_rate = bar_data.query("Bar not in @less_than_3times_list")['Y'].value_counts(1)[1]*100
print('Acceptance Rate for those who went to a bar 3 or fewer times a month : (' + str(less_than_3times_accept_rate) + '%)')
print('Acceptance Rate for those who went to a bar more than 3 times a month : (' + str(more_than_3times_accept_rate) + '%)')
#accept_less_than_3times = bar_data.query("Bar in @less_than_3times_list")['Y'].value_counts(0)[1]
```

## Insight

ANS: People who frequent bars more than 3 times a months are **twice likely** to accept the Bar coupon than the ones who go there 3 times or less.

4. Compare the acceptance rate between drivers who go to a bar more than once a month and are over the age of 25 to the all others.  Is there a difference?


```python
bar_data['age'].value_counts().sort_values()
```

```python
less_than_once_list = ['never','less1']
age_over_25 = ['50plus','46','41','36','31','26']

bar_data['OnePlus_Bar_Visit_25plus_Age'] = (~(bar_data['Bar'].isin(less_than_once_list)) & bar_data['age'].isin(age_over_25))

more_than_once_over_25_accept_rate = bar_data.query("OnePlus_Bar_Visit_25plus_Age == True")['Y'].value_counts(1)[1]*100
others_accept_rate = bar_data.query("OnePlus_Bar_Visit_25plus_Age == False")['Y'].value_counts(1)[1]*100
print('Acceptance Rate for drivers who go to a bar more than once a month and over the age of 25 : (' + str(more_than_once_over_25_accept_rate) + '%)')
print('Acceptance Rate for Others : (' + str(others_accept_rate) + '%)')
```

## Insight

ANS: People who frequent bars more than 3 times a months are **twice likely** to accept the Bar coupon than the ones who go there 3 times or less.

5. Use the same process to compare the acceptance rate between drivers who go to bars more than once a month and had passengers that were not a kid and had occupations other than farming, fishing, or forestry.


```python
less_than_once_list = ['never','less1']
passanger = ['Kid(s)']
occupation = ['Farming Fishing & Forestry']

bar_data['OnePlus_Bar_Visit_NoKid_Passenger_Not_Farming'] = (~(bar_data['Bar'].isin(less_than_once_list)) & ~(bar_data['passanger'].isin(passanger)) & ~(bar_data['occupation'].isin(occupation)))

OnePlus_Bar_Visit_NoKid_Passenger_Not_Farming = bar_data.query("OnePlus_Bar_Visit_NoKid_Passenger_Not_Farming == True")['Y'].value_counts(1)[1]*100
others_accept_rate = bar_data.query("OnePlus_Bar_Visit_NoKid_Passenger_Not_Farming == False")['Y'].value_counts(1)[1]*100
print('Acceptance Rate for drivers who go to a bar more than once a month and over the age of 25 : (' + str(OnePlus_Bar_Visit_NoKid_Passenger_Not_Farming) + '%)')
print('Acceptance Rate for Others : (' + str(others_accept_rate) + '%)')
```

## Insight

ANS: People who are 25+ and frequent bars more than once a months are **more than twice likely** to accept the Bar coupon than the others.

6. Compare the acceptance rates between those drivers who:

- go to bars more than once a month, had passengers that were not a kid, and were not widowed *OR*
- go to bars more than once a month and are under the age of 30 *OR*
- go to cheap restaurants more than 4 times a month and income is less than 50K.



```python
less_than_once_list = ['never','less1']
passanger = ['Kid(s)']
maritalStatus = ['Widowed']
age_under_30 = ['below21','21','26']
more_than_four_list = ['4-8','gt8']
income = ['Less than $12500','$12500 - $24999','$25000 - $37499','$37500 - $49999']

bar_data['OnePlus_Bar_Visit_NoKid_Passenger_Not_Widowed'] = (~(bar_data['Bar'].isin(less_than_once_list)) & ~(bar_data['passanger'].isin(passanger)) & ~(bar_data['maritalStatus'].isin(maritalStatus)))
cohort_1 = bar_data.query("OnePlus_Bar_Visit_NoKid_Passenger_Not_Widowed == True")['Y'].value_counts(1)[1]*100

bar_data['OnePlus_Bar_Visit_Under_Thirty'] = (~(bar_data['Bar'].isin(less_than_once_list)) & bar_data['age'].isin(age_under_30))
cohort_2 = bar_data.query("OnePlus_Bar_Visit_Under_Thirty == True")['Y'].value_counts(1)[1]*100

bar_data['FourPlus_Cheap_Restaurant_Lessthan_50K'] = (bar_data['RestaurantLessThan20'].isin(more_than_four_list) & bar_data['income'].isin(income))
cohort_3 = bar_data.query("FourPlus_Cheap_Restaurant_Lessthan_50K == True")['Y'].value_counts(1)[1]*100


print('Acceptance Rate for Cohort 1 : (' + str(cohort_1) + '%)')
print('Acceptance Rate for Cohort 2 : (' + str(cohort_2) + '%)')
print('Acceptance Rate for Cohort 3 : (' + str(cohort_3) + '%)')
```

## Insight

ANS: Cohort 2 (People under thirty and frequent the bars more than once) has the highest acceptance rate closely followed by Cohort 1 (People who are not widowed and frequent the bars more than once and do not take their kids during the trip).

Income doesn't seem to have a big impact / correlation with accepting counts vs not.
